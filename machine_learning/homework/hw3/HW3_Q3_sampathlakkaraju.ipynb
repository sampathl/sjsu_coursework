```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm 
X1 = []
Y1 = []
X2 = []
Y2 = []
data1=[]
data2=[]
data=[]
y_details=[]

def input_generation(a):
    for i in range(a):
        X1.append(random.uniform(0,1))
        Y1.append(random.uniform(0,1))
        X2.append(random.uniform(-1,0))
        Y2.append(random.uniform(-1,0))

    #label the data
    data1 = [np.array([1,X1[i],Y1[i],1]) for i in range(a)]
    data2 = [np.array([1,X2[i],Y2[i],-1]) for i in range(a)]
    data = data1 + data2
    
    return

input_generation(3)
# generates only the data not the matrices for computation
x1 = [np.array([X1[i]]) for i in range(3)]
x2 = [np.array([X2[i]]) for i in range(3)]
x = np.asarray(x1 + x2)
y1 = [np.array([Y1[i]]) for i in range(3)]
y2 = [np.array([Y2[i]]) for i in range(3)]
y = np.asarray(y1 + y2)
y_details.append(min(y1))
y_details.append(max(y2))
y_details.append((y_details[0]+y_details[1])/2)

print("generated 3 sets of data")
```

```python
x_reshaped = x.reshape(x.shape[0])
y_reshaped = y.reshape(y.shape[0])
x_sort=x_reshaped.copy()
x_sort.tolist()
x_sort=sorted(x_sort)
a=random.uniform(-1,1)
sep=random.uniform(y_details[1],y_details[0])
xs = np.linspace(-1,1,10)
line_d = np.array([sep for i in range(len(xs))])
plt.plot(xs, line_d,'g',label='Random')
line_d = np.array([y_details[2] for i in range(len(xs))])
plt.plot(xs, line_d,'r',label='svm')
plt.plot(x_sort, np.sign(x_sort),c='y',label='sign')
plt.scatter(x, y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```
